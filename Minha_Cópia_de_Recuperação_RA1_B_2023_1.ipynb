# Questão 2
Considerando a função definida por: 
A função descontínua $f(x)$ é definida como:

$$
f(x) = 
\begin{cases} 
\sin(x) & \text{se } -10 \leq x < -3 \\
\cos(x) & \text{se } -3 \leq x < 0 \\
x^2 & \text{se } 0 \leq x < 3 \\
e^{x-10} & \text{se } 3 \leq x \leq 10
\end{cases}
$$

Trace o gráfico desta função, marque os pontos de descontinuidade, caso existam estes pontos, com um círculo vermelho, coloque o valor destes pontos. Não se esqueça de colocar uma legenda que permita o entendimento do seu gráfico indicando as curvas correspondentes a cada equação. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função f(x)
def f(x):
    if -10 <= x < -3:
        return np.sin(x)
    elif -3 <= x < 0:
        return np.cos(x)
    elif 0 <= x < 3:
        return x**2
    elif 3 <= x <= 10:
        return np.exp(x - 10)

# Valores de x para o gráfico
x = np.linspace(-10, 10, 1000)

# Valores correspondentes de y
y = np.array([f(val) for val in x])

# Encontrando os pontos de descontinuidade
discontinuities = []
for i in range(len(x) - 1):
    if f(x[i]) != f(x[i + 1]):
        discontinuities.append((x[i], f(x[i])))

# Traçando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x)')

# Marcando os pontos de descontinuidade
if len(discontinuities) > 0:
    for point in discontinuities:
        plt.plot(point[0], point[1], 'ro')
        #plt.annotate(f'{point[1]:.2f}', (point[0], point[1]), xytext=(-15, 15), textcoords='offset points', ha='center', fontsize=8, color='red')

# Configurando o gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x)')
plt.axvline(x=0, color='gray', linestyle='--')
plt.axhline(y=0, color='gray', linestyle='--')
#plt.legend()
#plt.grid(True)

# Exibindo o gráfico
plt.show()
```
