# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import math
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x * math.exp(-x**2) - (x - 1)

def metodo_biseccao(a, b, epsilon):
    while abs(b - a) > epsilon:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

# Encontrar a solução aproximada usando o método da bissecção
a = 1
b = 2
epsilon = 0.0001
solution = metodo_biseccao(a, b, epsilon)

print("Solução aproximada: x =", solution)


x = sp.Symbol('x')
f1 = x * sp.exp(-x**2)
f2 = x - 1

area = sp.integrate(f1, (x, 2, solution)) - sp.integrate(f2, (x, 2, solution))

print("Área entre as curvas: ", area.evalf())


import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 2, 1000)
y1 = x * np.exp(-x**2)
y2 = x - 1

# Criar a figura e os eixos
fig, ax = plt.subplots()

# Plotar as curvas
ax.plot(x, y1, label='y = x*e^(-x^2)')
ax.plot(x, y2, label='y = x - 1')

# Preencher a área entre as curvas com uma cor diferente
ax.fill_between(x, y1, y2, where=(y1 > y2), color='cyan', alpha=0.5)

# Configurar o fundo do gráfico com uma cor diferente
ax.set_facecolor('white')

# Configurar os rótulos dos eixos
ax.set_xlabel('x')
ax.set_ylabel('y')

# Configurar o título do gráfico
ax.set_title('Área entre as curvas')

# Adicionar uma legenda
ax.legend()

# Exibir o gráfico
plt.show()
```

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração.

##a)
$$\int_0^2 x^2+1 \space dx$$

##b)
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Definir a função original
x = sp.symbols('x')
f = x**2 + 1

# Integrar a função
integral = sp.integrate(f, x)

# Converter a expressão simbólica da integral em uma função numérica
integral_func = sp.lambdify(x, integral)

# Traçar as funções
x_vals = np.linspace(0, 2, 100)
f_vals = np.array([sp.N(f.subs(x, val)) for val in x_vals])
integral_vals = np.array([integral_func(val) for val in x_vals])

# Criar o gráfico
plt.plot(x_vals, f_vals, label='Função Original')
plt.plot(x_vals, integral_vals, label='Função Integral')
plt.axvline(x=0, color='red', linestyle='--', label='Limite Inferior')
plt.axvline(x=2, color='green', linestyle='--', label='Limite Superior')

# Mostrar resultado da conta


# Configurar o gráfico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Integral de x^2 + 1')
plt.legend()

# Exibir o gráfico
plt.show()
```

#Questão 5 (vale até 1 ponto)
Encontre o valor da constante C para a antiderivada $F(x)$ da função $f(x) = 3x^3 - 2x + 1$ de tal forma que $F(1)=0$. Apresente seus cálculos em latex e explique as decisões que tomou.

Para encontrar a constante C, precisamos integrar a função f(x) primeiro. Usando a regra da potência da integral, temos:

$$\int f(x) dx = \int (3x^3 - 2x + 1) dx = \frac{3}{4}x^4 - x^2 + x + C $$

Agora, podemos usar a informação de que F(1) = 0 para encontrar o valor de C:

$$ F(1) = \frac{3}{4}(1)^4 - (1)^2 + (1) + C = 0 $$

Simplificando a equação acima, temos:

$$ \frac{3}{4} - 1 + 1 + C = 0 $$

$$ C = \frac{1}{4} $$

Portanto, o valor da constante C para a antiderivada F(x) da função f(x) é 1/4.
